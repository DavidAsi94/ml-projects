# Definition of Algorithms

```python
from collections import defaultdict
```

```python
# Algorithms

# Mean Rating: Calculate the mean rating for each movie, order with the highest rating listed first, and submit the top n.

# Output example:
# [ (4.2000000000000001, '222: The Matrix'), 
#   (3.5999999966666622, '233: Alien'), 
#   (3.3343434322222223, '3434: The Silence of the Lambs'), 
#   (3.2222222222222222, '111: The Water Divider'), 
#   (3.0022332223330625, '333: , Robocop')]

def topMean(prefs, n=5):
    
  # TODO

    dict = defaultdict(list)
    scores = []
    
    for k, v in prefs.items():
        for i in v:
            dict[i].append(v[i])

    for m, r in dict.items():
        scores.append((sum(r) / len(r),m))

    scores.sort(reverse=True)
    
    return scores[0:n]

topMean(critics)
```

```python
# % of ratings r+: Calculate the percentage of ratings for each movie that are r or higher. Order with the highest percentage first, and submit the top n.

# Output example:
# [ (0.8, '112: The Matrix'), 
#   (0.6663732233233444, '232: Star Wars: Episode IV'), 
#   (0.6, '232: Blade Runner'), 
#   (0.4534343343344433, '111: Gladiator'), 
#   (0.4, '555: Alien')]

def topPerc(prefs,r=3,n=5):
    
  # TODO

    scores = []
    ratings = defaultdict(list)
    dict = defaultdict(list)
    
    for k, v in prefs.items():
        for i in v:
            dict[i].append(v[i])    
    
    for key, value in dict.items():
        
         for i in value:
                if i >= r:
                    ratings[key].append(i)
    
    for key, value in ratings.items():
        scores.append((len(value) / len(dict[key]),key))

    scores.sort(reverse=True)
    
    return scores[0:n]

topPerc(critics)
```

```python
# Rating Count: Count the number of ratings for each movie, order with the most number of ratings first, and submit the top n.
# Output example:
# [ (27.0, '1: Toy Story'), 
#   (24.0, '112: Independency Day'), 
#   (20.0, '334: Alien'), 
#   (15.0, '222: Gladiator'), 
#   (12.0, '2333: I Robot')]

def topCount(prefs,n=5):
  # TODO

  scores = []
  ratings = defaultdict(list)
  dict = defaultdict(list)
    
  for k, v in prefs.items():
    for i in v:
        dict[i].append(v[i])   

  for key, value in dict.items():
    ratings[key].append(i)
    scores.append((float(len(value)),key))

  scores.sort(reverse=True)

  return scores[0:n]

topCount(critics)
```

```python
# Top 5 Movies related: Calculate movies that most often occur with 
# other movie, Star Wars: Episode IV - A New Hope (1977) by defautl
# using the (x+y)/x method. In other words, for each movie, calculate the 
# percentage of the other movie raters who also rated that movie. Order with 
# the highest percentage first, and submit the top 5.
#
# Output example:
# [ (0.9444333333222223, '1: Star Wards, Episode VI'), 
#   (0.9111111111666667, '2210: Star Wars: Episode VI - Return of the Jedi'), 
#   (0.8833333333333333, '2593: Indiana Jones'), 
#   (0.7444444444444444, '1780: Total Recall'), 
#   (0.5999999666666666, '3916: I Robot')]

# Return the top n movies that are related to a specific one. 
# Calculate movies that most often occur with other movie.

from collections import Counter

def topOccur(prefs,x = '260: Star Wars: Episode IV - A New Hope (1977)', n=5):

  # TODO
  scores = []  
  store = defaultdict(list)   # Stores x and filters by occurences when x was watched, together with other movies the user watched

  for k, v in prefs.items():
        for m, r in v.items():
            if x in v:
                store[x].append(m)

  inter_scores = Counter(store[x]) # Stores count of each movie's appearance within store variable
  store_2 = defaultdict(list)      # Stores ratio of how many times other movies were watched when the user watched x (unformatted)

  for k, v in inter_scores.items():
    store_2[k].append((v / inter_scores[x]))
    scores.append((store_2[k][0],k))  # Stores ratio of how many times other movies were watched when the user watched x (formatted)

  scores.sort(reverse = True)
  scores.remove((1.0, x))
    
  return scores[0:n]

topOccur(critics)
```

### Define functions to load and transform the large datasets

```python
import pandas as pd

# Load movies data: ID, movie, genre

def load_datasets(movie_data, user_data, rating_data):
    global movies, ratings
    movies = pd.read_csv(movie_data, sep='::', names = ['ID','Movie','Genre'],encoding='latin-1')

# Load user data: ID, age, sex, occupation, zip_code
    u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
    users = pd.read_csv(user_data, sep='::', names=u_cols, encoding='latin-1')

# Load ratings data: User ID, Movie ID, rating, timestamp
    r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
    ratings = pd.read_csv(rating_data, sep='::', names=r_cols, encoding='latin-1')
```

```python
# Create one merged DataFrame
def create_dataset(movies, ratings):
    global movie_ratings
    movie_ratings = movies.merge(ratings,left_on = 'ID',right_on = 'movie_id')
    print('The shape of the dataset is:' + str(movie_ratings.shape))
```

```python
# Reshape dataframe so it has the same format as the initial dataset: User as index, movie as column
# Output: Sparse matrix with many missing values
def transform_dataset(movie_ratings):
    movie_mio = movie_ratings.pivot_table(index="user_id", columns="Movie", values="rating",aggfunc='mean')
    movie_mio.to_csv('./movie_mio.csv')
```
